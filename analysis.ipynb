### Analysis and summary statistic methods

To actually perform computations on data, you will implement the following methods in `analysis.py`, modeled off of common functions in the `pandas` python package (https://pandas.pydata.org/pandas-docs/stable/user_guide/computation.html) that we will adopt later in the semester:

- `set_data(data)`
- `min(headers, rows=[])`: Computes the minimum of each variable in `headers` in the data object. Possibly only in a subset of data samples (`rows`) if `rows` is not empty.
- `max(headers, rows=[])`: Computes the maximum of each variable in `headers` in the data object.
- `range(headers, rows=[])`: Computes the range [min, max] for each variable in `headers` in the data object.
- `mean(headers, rows=[])`: Computes the mean for each variable in `headers` in the data object. Equation for mean: $$\bar{x} = \frac{1}{N} \sum_{i=1}^N x_i$$ where $x_i$ is the $i^{th}$ data sample and $N$ is the total number of data samples in the dataset.
- `var(headers, rows=[])`: Computes the variance for each variable in `headers` in the data object. Equation for sample variance: $$s^2 = \frac{1}{N-1} \sum_{i=1}^N (x_i - \bar{x})^2$$ where $x_i$ is the $i^{th}$ data sample, $N$ is the total number of data samples in the dataset, $\bar{x}$ is the mean.
- `std(headers, rows=[])`: Computes the standard deviation for each variable in `headers` in the data object. Equation for sample standard deviation: $$s^2 = \sqrt{\frac{1}{N-1} \sum_{i=1}^N (x_i - \bar{x})^2}$$ where $x_i$ is the $i^{th}$ data sample, $N$ is the total number of data samples in the dataset, $\bar{x}$ is the mean.

***Data Analysis***

This data set contains various metrics measured at time of a forest fire (source: https://archive.ics.uci.edu/ml/datasets/Forest+Fires). The data was compiled to be used with machine learning in an an attempt to predict how much of a forest in northeast Portugal would be burned during a fire. There are many metrics including the FFMC index (the fine fuel moisture code, which represents fuel moisture of forest litter fuels under the shade of a forest canopy), the DMC index (Duff moisture code, represents fuel moisture of decmomposed organic material underneath the litter), DC index (drought code, approximates moisture conditions for fuels), ISI (initial spread index, a fire behavior index), temperature (celsius), relative humidity, wind speed (in km/h), rain (in mm/m^2), and area burned (in ha). I divided the data into two groups (indices and real measures), discarding area beacause the source notes that area is greatly skewed. Then, I plotted the first group and the second group separately using pair plot. I thought that there may be a correlation between temperature and wind speed, and was curious to see how the various FWI indices were related to each other. 

From the scatter plots of the indices, it appears as though the strongest correlation occurs between DMC and DC. This is fairly sensible as the DMC measures decomposed organic material on the ground and the DC measures drying deep into the soil. Thus, if the DC is low, likely the DMC is very low. None of the physical measures have particularly strong correlations based on the set of scatter plots. Out of curiousity, I plotted wind and drought code against temperature individually using the scatter() method. I had thought higher temperature and high wind may correlate in forest fire conditions, or DC and temperature, but there was no evident correlation between either of these variables. 

The lack of overwhelming correlation bewteen any two variables indicates that the conditions of of a forest fire are complicated and originate from complex interactions between a variety of variables. As this dataset was with various machine learning techniques for analysis, it is unsurprisng that simple attempts to divine linear correlations fall short.
